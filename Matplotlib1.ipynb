```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)#Find the sine value of all values of x
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

#### Plotting the data

```python
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding a title to the x-axis and y_axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('Sine Value')
plt.title('Sine Function')
plt.show()
```

```python
# Parabola plotting
a = np.linspace(-10,10,20)
b = a**2


plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,b,'r+')
plt.show()
```

```python
plt.plot(x,y, 'g.')
plt.show()
```

```python
plt.plot(a,b,'rx')
#Plots in red colours using 'x' symbol.
```

```python
c= np.linspace(-5,5,50)
plt.plot(c, np.sin(c),'g-')  #Green lineplot
plt.plot(c,np.cos(c),'r--')   #Red dashed line plot

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #Plots from origi (0,0) to a maximum value (1,1)
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel("Number of people")
plt.show()
```

#### Pie Chart

```python
fig1 = plt.figure()
ax1= fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax1.pie(people, labels=languages, autopct = '%1.1f%%')

plt.show()
```

#### Scatter Plot

```python
e = np.linspace(0,10,30)
f = np.sin(e)
g = np.cos(e)
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y, color='g')
ax2.scatter(x,z,color='b')
plt.show()
```

```python
fig3 = plt.figure()
ax3 = plt.axes(projection = '3d')
i = 20 *  np.random.random(100)
j = np.sin(i)
k = np.cos(i)
ax3.scatter(i,j,k,c=k, cmap='Blues')
plt.show()
```
