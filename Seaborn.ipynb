```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Some libraries come with in-built datasets
sns.get_dataset_names()
```

```python
df = sns.load_dataset('mpg')
df
```

```python
df.info()
```

# 1. Distribution Plots

```python
#`distplot` is a deprecated function and will be removed in seaborn v0.14.0.

# Please adapt your code to use either `displot` (a figure-level function with
# similar flexibility) or `histplot` (an axes-level function for histograms).
#sns.distplot(df['mpg'])

sns.displot(df['mpg'])
plt.show()
```

```python
sns.histplot(df['mpg'], kde=False, color='red', bins=20)
plt.show()
```

```python
df
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg',y='horsepower', data=df, hue='cylinders', size = 'weight')
plt.show()
```

```python
sns.jointplot(x='weight', y='mpg', data=df, kind='hex')
#Helps you to visulaise histograms of each variable and then the scatterplot of both.
```

```python
sns.jointplot(x='weight', y='mpg', data=df, kind='reg')
```

#### pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns)

```python
df.head(5)
```

```python
df_subset = df[['mpg','cylinders','weight']]
```

```python
df_subset
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df_subset)
plt.show()
```

```python
df_subset
```

```python
sns.pairplot(df_subset, hue='cylinders', palette='coolwarm')
#huues is used for categorical variables.
#palette specifies the colour.
```

# 2. Categorical plots

```python
sns.boxplot(x='cylinders',y='mpg', data=df)
```

```python
sns.barplot(x='cylinders',y='mpg',data= df)
#barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean.
#The black vertical lines are the error bars.
```

```python
sns.countplot(x='cylinders', data= df)
```

# 3. Heat Map

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),cmap='coolwarm', annot = True)
plt.show()
```

# 4. Facet Grids

```python
df.head()
```

```python
df.origin.value_counts()
```

```python
g = sns.FacetGrid(df, col='cylinders', row='origin')
g = g.map(plt.scatter, 'mpg','weight') #Apply a plotting function to each facet's subset of the data
```

```python
g = sns.FacetGrid(df, col='cylinders', hue='origin')
g = g.map(plt.scatter, 'mpg','weight').add_legend()
```
